```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

from Marchenko_depthloop import Marchenko_depthloop
from Marchenko_depthloop_fullsurvey import Marchenko_depthloop_fullsurvey
```

```python
FUP1, FDOWN1, PUP1_full, PDOWN1_full, PUP2_full, PDOWN2_full, redatumed1_full, redatumed2_full = \
    Marchenko_depthloop_fullsurvey(900, 900)
```

```python
clip=0.2
fig, axs = plt.subplots(1, 2, sharey=True, figsize=(10, 7))
axs[0].imshow(FDOWN1[:,45].T, cmap='gray', 
              vmin=-clip*np.abs(FDOWN1).max(), vmax=clip*np.abs(FDOWN1).max())
axs[0].set_title(r'$F^+_1$')
axs[0].axis('tight')
axs[1].imshow(FUP1[:,45].T, cmap='gray', 
              vmin=-clip*np.abs(FDOWN1).max(), vmax=clip*np.abs(FDOWN1).max())
axs[1].set_title(r'$F^-_1$')
axs[1].axis('tight')
axs[1].set_ylim(1500, 500)

clip=0.2
fig, axs = plt.subplots(1, 4, sharey=True, figsize=(20, 7))
axs[0].imshow(PDOWN1_full[:,45].T, cmap='gray', 
              vmin=-clip*np.abs(PDOWN1_full).max(), vmax=clip*np.abs(PDOWN1_full).max())
axs[0].set_title(r'$G^+_1$')
axs[0].axis('tight')
axs[1].imshow(PUP1_full[:,45].T, cmap='gray', 
              vmin=-clip*np.abs(PDOWN1_full).max(), vmax=clip*np.abs(PDOWN1_full).max())
axs[1].set_title(r'$G^-_1$')
axs[1].axis('tight')
axs[2].imshow(PDOWN2_full[:,45].T, cmap='gray', 
              vmin=-clip*np.abs(PDOWN1_full).max(), vmax=clip*np.abs(PDOWN1_full).max())
axs[2].set_title(r'$G^+_2$')
axs[2].axis('tight')
axs[3].imshow(PUP2_full[:,45].T, cmap='gray', 
              vmin=-clip*np.abs(PDOWN1_full).max(), vmax=clip*np.abs(PDOWN1_full).max())
axs[3].set_title(r'$G^-_2$')
axs[3].axis('tight')
axs[3].set_ylim(600, 0);
plt.savefig('Marchenko_fields_full', dpi=300)

clip=0.1
fig, axs = plt.subplots(1, 3, sharey=True, figsize=(10, 7))
axs[0].imshow(redatumed1_full[:,45].T, cmap='gray', 
              vmin=-clip*np.abs(redatumed1_full).max(), vmax=clip*np.abs(redatumed1_full).max())
axs[0].set_title(r'$R_1$')
axs[0].axis('tight')
axs[1].imshow(redatumed2_full[:,45].T, cmap='gray', 
              vmin=-clip*np.abs(redatumed1_full).max(), vmax=clip*np.abs(redatumed1_full).max())
axs[1].set_title(r'$R_2$')
axs[1].axis('tight')
axs[2].imshow(redatumed1_full[:,45].T-redatumed2_full[:,45].T, cmap='gray', 
              vmin=-clip*np.abs(redatumed1_full).max(), vmax=clip*np.abs(redatumed1_full).max())
axs[2].set_title(r'$dR$')
axs[2].axis('tight')
axs[2].set_ylim(800, 400);
plt.savefig('MDD_fields_full', dpi=300)
```
