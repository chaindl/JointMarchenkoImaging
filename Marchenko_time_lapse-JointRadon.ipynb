```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pylops

from Marchenko_depthloop_JointRadon import Marchenko_depthloop_JointRadon
```

```python
FUP1, FDOWN1, PUP1, PDOWN1, PUP2, PDOWN2, redatumed1, redatumed2 = \
    Marchenko_depthloop_JointRadon(900, 900, 1)
```

```python
ivs = 45

clip=0.2
fig, axs = plt.subplots(1, 2, sharey=True, figsize=(10, 7))
axs[0].imshow(FDOWN1[:,ivs].T, cmap='gray', 
              vmin=-clip*np.abs(PDOWN1).max(), vmax=clip*np.abs(PDOWN1).max())
axs[0].set_title(r'$F^+_1$')
axs[0].axis('tight')
axs[1].imshow(FUP1[:,ivs].T, cmap='gray', 
              vmin=-clip*np.abs(PDOWN1).max(), vmax=clip*np.abs(PDOWN1).max())
axs[1].set_title(r'$F^-_1$')
axs[1].axis('tight')
axs[1].set_ylim(1500, 500)

fig, axs = plt.subplots(1, 4, sharey=True, figsize=(20, 7))
axs[0].imshow(PDOWN1[:,ivs].T, cmap='gray', 
              vmin=-clip*np.abs(PDOWN1).max(), vmax=clip*np.abs(PDOWN1).max())
axs[0].set_title(r'$G^+_1$')
axs[0].axis('tight')
axs[1].imshow(PUP1[:,ivs].T, cmap='gray', 
              vmin=-clip*np.abs(PDOWN1).max(), vmax=clip*np.abs(PDOWN1).max())
axs[1].set_title(r'$G^-_1$')
axs[1].axis('tight')
axs[2].imshow(PDOWN2[:,ivs].T, cmap='gray', 
              vmin=-clip*np.abs(PDOWN1).max(), vmax=clip*np.abs(PDOWN1).max())
axs[2].set_title(r'$G^+_2$')
axs[2].axis('tight')
axs[3].imshow(PUP2[:,ivs].T, cmap='gray', 
              vmin=-clip*np.abs(PDOWN1).max(), vmax=clip*np.abs(PDOWN1).max())
axs[3].set_title(r'$G^-_2$')
axs[3].axis('tight')
axs[3].set_ylim(600, 0);
```

```python
clip=0.1
fig, axs = plt.subplots(1, 3, sharey=True, figsize=(10, 7))
axs[0].imshow(redatumed1[:,45].T, cmap='gray', 
              vmin=-clip*np.abs(redatumed1).max(), vmax=clip*np.abs(redatumed1).max())
axs[0].set_title(r'$R_1$')
axs[0].axis('tight')
axs[1].imshow(redatumed2[:,45].T, cmap='gray', 
              vmin=-clip*np.abs(redatumed1).max(), vmax=clip*np.abs(redatumed1).max())
axs[1].set_title(r'$R_2$')
axs[1].axis('tight')
axs[2].imshow(redatumed1[:,45].T-redatumed2[:,45].T, cmap='gray', 
              vmin=-clip*np.abs(redatumed1).max(), vmax=clip*np.abs(redatumed1).max())
axs[2].set_title(r'$dR$')
axs[2].axis('tight')
axs[2].set_ylim(800, 400);
plt.savefig('MDD_fields_joint.png', dpi=300)
```
